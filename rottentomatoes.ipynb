```python
# Dependencies
import requests
from bs4 import BeautifulSoup as bs
from splinter import Browser
import pandas as pd
```

```python
# Create path and browser
executable_path={'executable_path': 'chromedriver.exe'}
browser=Browser('chrome', **executable_path, headless=False)

# URL of page to be scraped
url="https://editorial.rottentomatoes.com/guide/best-netflix-movies-to-watch-right-now/"
browser.visit(url)
```

```python
# Loop the results to find title, year, rating
dict={}
mov=[]
for x in range(3):
    html = browser.html
    soup=bs(html, 'html.parser')
    movies=soup.find_all('div',class_="article_movie_title")
    
    for movie in movies:
        title=movie.find('a').text
        year=movie.find('span',class_='subtle start-year').text
        year=int(year[1:5])
        rating=movie.find('span',class_='tMeterScore').text
        rating=float(rating[:2])/10
        dict={'title':title,
             'year':year,
             'rating':rating}
        mov.append(dict)
        print(dict)
        try:
            browser.find_by_name('  Next ').click()
        except:
            pass
```

```python
# Create a dataframe and export:
df=pd.DataFrame(mov)
#df.to_csv('df.csv')
df
```
